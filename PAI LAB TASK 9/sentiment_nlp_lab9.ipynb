###Sentiment Analysis with Basic NLP
Tokenization + Stopwords Removal + Sentiment Analysis

```python
from nltk.corpus import stopwords
from textblob import TextBlob
import nltk
import re

nltk.download('stopwords')

my_text = "I had a wonderful dream but it was also a bit scary"

words = re.findall(r'\w+', my_text)
print("Words:", words)

stop_words = set(stopwords.words('english'))

clean_words = []
for w in words:
    if w.lower() not in stop_words:
        clean_words.append(w)

print("After cleaning:", clean_words)

final_text = " ".join(clean_words)

tb = TextBlob(final_text)
score = tb.sentiment.polarity

if score > 0:
    result = "Positive"
elif score < 0:
    result = "Negative"
else:
    result = "Neutral"

print("Final Text:", final_text)
print("Score:", score)
print("Result:", result)
```
