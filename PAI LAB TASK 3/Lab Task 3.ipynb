```python
# Water Jug Problem using DFS

jug1_capacity = 4
jug2_capacity = 3
goal = 2

visited = set()

def dfs(jug1, jug2, path):

    if (jug1, jug2) in visited:
        return False

    visited.add((jug1, jug2))

    path.append((jug1, jug2))

    # Goal condition
    if jug1 == goal:
        print("Solution Found\n")
        print("Steps:")
        for step in path:
            print(step)
        return True

    # Rule 1: Fill Jug1
    if jug1 < jug1_capacity:
        print("Rule: Fill Jug1")
        if dfs(jug1_capacity, jug2, path.copy()):
            return True

    # Rule 2: Fill Jug2
    if jug2 < jug2_capacity:
        print("Rule: Fill Jug2")
        if dfs(jug1, jug2_capacity, path.copy()):
            return True

    # Rule 3: Empty Jug1
    if jug1 > 0:
        print("Rule: Empty Jug1")
        if dfs(0, jug2, path.copy()):
            return True

    # Rule 4: Empty Jug2
    if jug2 > 0:
        print("Rule: Empty Jug2")
        if dfs(jug1, 0, path.copy()):
            return True

    # Rule 5: Pour Jug1 -> Jug2
    if jug1 > 0 and jug2 < jug2_capacity:
        transfer = min(jug1, jug2_capacity - jug2)
        print("Rule: Pour Jug1 -> Jug2")
        if dfs(jug1 - transfer, jug2 + transfer, path.copy()):
            return True

    # Rule 6: Pour Jug2 -> Jug1
    if jug2 > 0 and jug1 < jug1_capacity:
        transfer = min(jug2, jug1_capacity - jug1)
        print("Rule: Pour Jug2 -> Jug1")
        if dfs(jug1 + transfer, jug2 - transfer, path.copy()):
            return True

    return False


# Start from (0,0)
dfs(0, 0, [])
```
